# Run profile

## Single scenarios

A single scenario is defined as a magnitude/percentile/submodel. Here is an example (note all parameters must be passed as keyword arguments).

```python
mag, perc, model = 6.8, 0.84, "d_md"

results = run_displacement_profile.run_profile(
    magnitude=mag, percentile=perc, submodel=model
)

results.head()
```

The default location spacing is 0.05, but it can be changed with `location_step=0.01`, for example:

```python
results = run_displacement_profile.run_profile(
    magnitude=mag, percentile=perc, submodel=model, location_step=0.01
)

results.head()
```

And the profile can be plotted:

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))

label = f"{model} (w/o magnitude aleatory var.)"
ax.plot(results["location"], results["displ_without_aleatory"], label=label)

label = f"{model} (w/ magnitude aleatory var.)"
ax.plot(results["location"], results["displ_with_aleatory"], ls=":", label=label)

ptile = "Mean" if perc == -1 else f"P({int(perc*100)})"
title = f"Magnitude {mag}, Percentile={ptile}"
ax.set(xlabel="x/L", ylabel="Principal Displacement (m)", title=title)
ax.legend(loc="best");
```

This model should only be used for reverse faulting. That is the default style. A `UserWarning` will be issued and the user input will be over-ridden in the output table if another style is entered, such as `style="strike-slip"`:

```python
results = run_displacement_profile.run_profile(
    magnitude=mag, percentile=perc, submodel=model, style="strike-slip"
)

results.head()
```

The Moss and Ross (2011) model provides two alternative submodels: one is based on normalized displacement $D/AD$ and the other is $D/MD$. The user must specify which model is used, and an error results if an invalid model is entered.

```python
try:
    results = run_displacement_profile.run_profile(
        magnitude=mag, percentile=perc, submodel="meow"
    )
except ValueError as e:
    print(e)
```

```python
del mag, perc, model
```

## Multiple scenarios

Multiple magnitude/percentile/submodel scenarios can be run. All combinations of magnitudes, percentiles, and submodels are run. Here is an example.

```python
mag, perc, model = 7, [0.05, 0.5, 0.95, -1], ["D_AD", "D_MD"]

results = run_displacement_profile.run_profile(
    magnitude=mag, percentile=perc, submodel=model
)
```

And we can plot the different scenarios:

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 7))

lines = {0.05: ":", 0.95: ":", 0.5: "--", -1: "-"}
axs = {"d_ad": axes[0], "d_md": axes[1]}

for scenario, group in results.groupby(["model_name", "percentile"]):
    group = group.sort_values(by="location")
    _model, _perc = scenario[0], scenario[1]
    ptile = "Mean" if _perc == -1 else f"P({int(_perc*100)})"
    label = f"{_model.upper()}, {ptile}"

    axs[_model][0].plot(
        group["location"], group["d_xd"], c="k", ls=lines[_perc], label=label
    )

    axs[_model][1].plot(
        group["location"],
        group["displ_without_aleatory"],
        c="k",
        ls=lines[_perc],
        label=label,
    )

    axs[_model][2].plot(
        group["location"],
        group["displ_with_aleatory"],
        c="k",
        ls=lines[_perc],
        label=label,
    )

for _ax in axes.flatten():
    _ax.set(xlabel="x/L", xlim=[0, 0.5], ylim=[0, 5])
    _ax.legend(loc="upper left")


axes[0][0].set(ylabel="D/AD", title="Normalized Displacment")
axes[1][0].set(ylabel="D/MD", title="Normalized Displacment", ylim=[0, 1])
axes[0][1].set(ylabel="Displ. (m)", title="Partial Aleatory")
axes[0][2].set(ylabel="Displ. (m)", title="Full Aleatory")
axes[1][1].set(ylabel="Displ. (m)", title="Partial Aleatory")
axes[1][2].set(ylabel="Displ. (m)", title="Full Aleatory")

fig.subplots_adjust(hspace=0.4, wspace=0.2)
fig.suptitle(f"Magnitude {mag}", fontsize=14)

plt.savefig(f"scratch-mr11 slip profiles with quantiles.png", bbox_inches="tight")
```

### Investigating why D/MD mean is less than median for ~ x/L=0.5 with just location aleatory

```python
from scipy import stats
from functions import _calc_distrib_params_d_ad, _calc_distrib_params_d_md
```

```python
xl = 0.5

# Results for D/AD
_alpha, _beta = _calc_distrib_params_d_ad(location=xl)
d_ad_test_arr = np.arange(0, 5, 0.01)
density_d_ad = stats.gamma.pdf(x=d_ad_test_arr, a=_alpha, scale=_beta, loc=0)
cdf_d_ad = stats.gamma.cdf(x=d_ad_test_arr, a=_alpha, scale=_beta, loc=0)
# d_ad_samples = stats.gamma.rvs(a=_alpha, scale=_beta, loc=0, size=500000)
print(_alpha, _beta)

# Results for D/MD
_alpha, _beta = _calc_distrib_params_d_md(location=xl)
# d_md_test_arr = np.arange(0.01, 0.92, 0.01) # Robb's range in XLS
d_md_test_arr = np.arange(0.0, 1.01, 0.01)
density_d_md = stats.beta.pdf(x=d_md_test_arr, a=_alpha, b=_beta)
cdf_d_md = stats.beta.cdf(x=d_md_test_arr, a=_alpha, b=_beta)
# d_md_samples = stats.beta.rvs(a=_alpha, b=_beta, size=500000)
print(_alpha, _beta)

fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0][0].plot(d_ad_test_arr, density_d_ad, c="k")
ax[0][0].set(xlabel="D/AD", ylabel="Probability Density")
ax[0][1].plot(d_md_test_arr, density_d_md, c="k")
ax[0][1].set(xlabel="D/MD", ylabel="Probability Density")

ax[1][0].plot(d_ad_test_arr, cdf_d_ad, c="k")
ax[1][0].set(xlabel="D/AD", ylabel="Cumulative Dist.")
ax[1][1].plot(d_md_test_arr, cdf_d_md, c="k")
ax[1][1].set(xlabel="D/AD", ylabel="Cumulative Dist.")
ax[1][0].axhline(y=0.5, color="black", lw=0.5, label=None)
ax[1][1].axhline(y=0.5, color="black", lw=0.5, label=None)


ax[1][0].text(2.1, 0.55, "More mass above median,\nso mean skews high.")
ax[1][1].text(0.0, 0.35, "More mass below median,\nso mean skews low.")


fig.suptitle(
    f"Aleatory Variability Models for x/L= {xl}\n(Curves verified with calc sheet from R. Moss 8/2/2021)",
    y=0.95,
    fontsize=14,
)

plt.savefig(f"scratch-mr11 pdfs for xl {str(xl)}.png", bbox_inches="tight")
```
