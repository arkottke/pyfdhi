## Loop for multiple scenarios

```python
# User-defined cases
magnitude, location, style, percentile = (
    [6.8, 7.4],
    [0.25, 0.5],
    ["reverse", "normal"],
    [0.5, -1],
)

# Define all scenarios
if not isinstance(magnitude, list):
    magnitude = [magnitude]
if not isinstance(location, list):
    location = [location]
if not isinstance(percentile, list):
    percentile = [percentile]
if not isinstance(style, list):
    style = [style]

# Perform calculations
scenarios = list(product(magnitude, location, style, percentile))
df = pd.DataFrame()
for scenario in scenarios:
    results = run_displacement_model.run_model(
        magnitude=scenario[0],
        location=scenario[1],
        style=scenario[2],
        percentile=scenario[3],
    )
    df = pd.concat([df, results])

# Clean up variables
del magnitude, location, style, percentile

df
```

# Calculate slip profile

```python
# User-defined cases
location = np.arange(0, 1.01, 0.01).tolist()
magnitude = 7
percentile = -1
style = "strike-slip"

# Define all scenarios
if not isinstance(magnitude, list):
    magnitude = [magnitude]
if not isinstance(location, list):
    location = [location]
if not isinstance(percentile, list):
    percentile = [percentile]
if not isinstance(style, list):
    style = [style]

# Perform calculations
scenarios = list(product(magnitude, location, style, percentile))
df = pd.DataFrame()
for scenario in scenarios:
    results = run_displacement_model.run_model(
        magnitude=scenario[0],
        location=scenario[1],
        style=scenario[2],
        percentile=scenario[3],
    )
    df = pd.concat([df, results])

df
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
x, y_left, y_right, y_folded = (
    df["location"],
    df["displ_site"],
    df["displ_complement"],
    df["displ_folded"],
)
ax.plot(x, y_left, c="tab:orange", label="left")
ax.plot(x, y_right, c="tab:blue", label="folded")
ax.plot(x, y_folded, c="k", label="right")
title = f"Magnitude {magnitude}, {style}, Percentile={percentile}"
ax.set(xlabel="U*", ylabel="Aggregated Displacement (m)", title=title)
ax.legend(loc="best")

# Clean up variables
del magnitude, location, style, percentile
```
